```python
#global
df['bet_mean_global'] = df.groupby('user_id')['fiat_bet_amount'].transform('mean')


df["bet_mean_weighted_a001_ns"] = (
    df.groupby("user_id")["fiat_bet_amount"]
      .transform(lambda x: x.ewm(alpha=0.01, adjust=False).mean())
)



past_bets = df.groupby("user_id")["fiat_bet_amount"].shift(1)

# df["bet_mean_weighted_a01"] = (
#     past_bets
#     .groupby(df["user_id"])
#     .transform(lambda s: s.ewm(alpha=0.1, adjust=False).mean())
# )
# df["bet_mean_weighted_a01"] = df["bet_mean_weighted_a01"].fillna(
#     df.groupby("user_id")["fiat_bet_amount"].transform("first")
# )


df["bet_mean_weighted_a001"] = (
    past_bets
    .groupby(df["user_id"])
    .transform(lambda s: s.ewm(alpha=0.01, adjust=False).mean())
)
df["bet_mean_weighted_a001"] = df["bet_mean_weighted_a001"].fillna(
    df.groupby("user_id")["fiat_bet_amount"].transform("first")
)


```

```python
df['bet_proportion_global'] = df['fiat_bet_amount'] / df['bet_mean_global']
# df["bet_proportion_weighted_a01"] = df["fiat_bet_amount"] / df["bet_mean_weighted_a01"]
df["bet_proportion_weighted_a001"] = df["fiat_bet_amount"] / df["bet_mean_weighted_a001"]
df["bet_proportion_weighted_a001_ns"] = df["fiat_bet_amount"] / df["bet_mean_weighted_a001_ns"]

```

```python
df.head(20)
```

```python
df['risk_global'] = df['odds_adjusted'] * df['bet_proportion_global']
df['risk_weighted_a001_ns'] = np.log1p(df['odds_adjusted'] * df['bet_proportion_weighted_a001_ns'])

eps = 1e-9  
odds = df["odds_adjusted"].clip(lower=1) 
prop = df["bet_proportion_weighted_a001"].clip(lower=eps)  


def risk_log_gamma_delta(gamma=1.0, delta=1.0):
    return gamma * np.log(odds) + delta * np.log(prop)

# exemplos:
df["risk_g5d5"] = risk_log_gamma_delta(0.5, 0.5)
df["risk_g6d4"] = risk_log_gamma_delta(0.6, 0.4)
df["risk_g7d3"] = risk_log_gamma_delta(0.7, 0.3)
```

```python
def sci_br(x, digits=6, sci_threshold=1e-4, math=True):
    if not np.isfinite(x):
        return ""
    if x == 0:
        return "0"
    ax = abs(x)
    if ax < sci_threshold:
        s = f"{x:.{digits}e}"          # ex: '1.234567e-04'
        mant, exp = s.split("e")
        mant = mant.replace(".", ",")
        exp = int(exp)
        return rf"${mant}\times 10^{{{exp}}}$" if math else f"{mant}e{exp}"
    else:
        s = f"{x:.{digits}f}".rstrip("0").rstrip(".")
        return s.replace(".", ",")
```

```python

def downcast_df(df: pd.DataFrame) -> pd.DataFrame:
    float_cols = df.select_dtypes(include=['float64','float32']).columns
    df[float_cols] = df[float_cols].astype('float32', copy=False)

    int_cols = df.select_dtypes(include=['int64','int32','int']).columns
    for c in int_cols:
        if df[c].min() >= 0 and df[c].max() <= np.iinfo(np.uint32).max:
            df[c] = df[c].astype('uint32', copy=False)
        else:
            df[c] = df[c].astype('int32', copy=False)


    if 'user_id' in df.columns:
        df['user_id'] = df['user_id'].astype('category')


    if 'game_id' in df.columns:
        df['game_id'] = df['game_id'].astype('category')

    gc.collect()
    return df
```
